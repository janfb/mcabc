```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from scipy.stats import gamma, beta
import scipy
from scipy.special import gammaln, betaln
```

```python
%matplotlib inline

mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['axes.labelsize'] = 15
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 15
mpl.rcParams['figure.figsize'] = (15, 5)
```

## Gamma distribution

```python
thetas = np.linspace(0, 30, 1000)
ks = [1., 2., 3., 5., 9., 7.5, 0.5, 5., 1., 5.]
ths = [2., 2., 2., 1., 0.5, 1., 1., 2., 1., 7.2]

for k, th in zip(ks, ths):
    plt.plot(thetas, gamma.pdf(x=thetas, a=k, scale=th), label=r'k={}, $\theta$={}'.format(k, th))
plt.ylim([0., 0.5])
plt.legend()
plt.title('Gamma distribution priors');
```

## Beta distribution

```python
x = np.linspace(0, 1., 100)
alp = [0.5, 1., 2., 2., 5., 5.]
bet = [0.5, 3., 2., 5., 1., 2.]

for a, b in zip(alp, bet): 
    plt.plot(x, beta.pdf(x, a, b), label='a={}, b={}'.format(a, b))

plt.legend();
```

## Prior on p in Poisson-Gamma mixture 

```python
ps = np.linspace(1e-8, 1 - 1e-8, 1000)

p_pdf = lambda p, k, th: np.power(1-p, -2) * gamma.pdf(p / (1-p), a=k, scale=th)

k = 2.
thetas = np.logspace(-2, 1, 50)

for th in thetas: 
    plt.plot(ps, p_pdf(ps, k, th))
```
